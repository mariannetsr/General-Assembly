## Feature Engineering

Load cleaned datasets:

```python
train = pd.read_csv('../datasets/train_cleaned.csv')

test = pd.read_csv('../datasets/test_cleaned.csv')
```

### Next steps:

- Convert ordinal categories to numerical variables e.g.`bsmt_cond`, `bsmt_qual`


- Combine related variables like fireplace variables and garage variables to create a score e.g. `df['fireplace_score'] = df['fireplaces']* df['fireplace_qu']` and drop the irrelevant ones

- Create dummy variables (one-hot encoding)

```python
# ------------------------------ CREATE FUNCTION TO CONVERT VALUES ------------------------------
    
# defining dictionaries for renaming

# ms_subclass_di = {20: '1-STORY_NEW', 
#                   30: '1-STORY_OLD', 
#                   40: '1-STORY_FINISHED_ATTIC', 
#                   45: '1-1/2_STORY_UNFINISHED_ALL',
#                   50: '1-1/2_STORY_FINISHED_ALL', 
#                   60: '2-STORY_NEW',
#                   70: '2-STORY_OLD', 
#                   75: '2-1/2_STORY_ALL',
#                   80: 'SPLIT-LEVEL', 
#                   85: 'SPLIT-FOYER',
#                   90: 'DUPLEX-ALL', 
#                   120: '1-STORY_PUD',
#                   150: '1-1/2_STORY_PUD',
#                   160: '2-STORY_PUD',
#                   180: 'MULTI_PUD',
#                   190: '2_FAM_CONV_ALL'}

lotshape_di = {'Reg': 0,
                'IR1': 1,
                'IR2': 2,
                'IR3': 3}

bsmtex_di = {'None': 0,
            'No': 1,
            'Mn': 2,
            'Av': 3,
            'Gd': 4}

bsmtfin_di = {'GLQ': 6,
              'ALQ': 5,
              'BLQ': 4,
              'Rec': 3,
              'LwQ': 2,
              'Unf': 1,
              'None': 0}

qual_di = {'Ex': 5,
            'Gd': 4,
            'TA': 3,
            'Fa': 2,
            'Po': 1,
            'None': 0}

util_di = {'AllPub': 4, 
           'NoSewr' : 3, 
           'NoSeWa': 2, 
           'ELO' : 1}

landslope_di = {'Gtl': 3, 
                'Mod' : 2, 
                'Sev': 1}

garagefin_di = {'None': 0, 
                'Unf': 1, 
                'RFn' : 2, 
                'Fin': 3}

    
def convert_values(df):
    
    df["mo_sold"] = df["mo_sold"].astype(str)
    
#    df["ms_subclass"] = df["ms_subclass"].map(ms_subclass_di)
   
    df["lot_shape"] = df["lot_shape"].map(lotshape_di)
    
    df["utilities"] = df["utilities"].map(util_di)              
                                          
    df["land_slope"] = df["land_slope"].map(landslope_di)
    
    df["exter_qual"] = df["exter_qual"].map(qual_di)
    df["exter_cond"] = df["exter_cond"].map(qual_di)
    
    df["bsmt_qual"] = df["bsmt_qual"].map(qual_di)
    df["bsmt_cond"] = df["bsmt_cond"].map(qual_di)
    df["bsmt_exposure"] = df["bsmt_exposure"].map(bsmtex_di)
    df["bsmtfin_type_1"] = df["bsmtfin_type_1"].map(bsmtfin_di)
    df["bsmtfin_type_2"] = df["bsmtfin_type_2"].map(bsmtfin_di)

    df["heating_qc"] = df["heating_qc"].map(qual_di)
        
    df["kitchen_qual"] = df["kitchen_qual"].map(qual_di)
    
    df["fireplace_qu"] = df["fireplace_qu"].map(qual_di)
    
    df["garage_finish"] = df["garage_finish"].map(garagefin_di)
    df["garage_qual"] = df["garage_qual"].map(qual_di)
    df["garage_cond"] = df["garage_cond"].map(qual_di)
    
    df["paved_drive"] = df["paved_drive"].map({"N": 1, "P": 2, "Y" : 3})
    
    df["pool_qc"] = df["pool_qc"].map(qual_di)
```

```python
# check if total_bsmt_sf is the sum of the 3 other bsmt_sf variables

train['total_bsmt_sf'].sum()-train['bsmtfin_sf_1'].sum()\
- train['bsmtfin_sf_2'].sum()-train['bsmt_unf_sf'].sum()
```

Confirmed that `total_bsmt_sf` is the sum of the 3 other `bsmt_sf` variables. We can drop those 3 columns.

```python
# ------------------------------ CREATE FUNCTION TO COMBINE AND DROP COLUMNS ------------------------------

def combine_and_drop(df):
    
    # create a total pool score
    df["pool_score"] = df["pool_area"]*df["pool_qc"]
    
    # create a total garage score
    df["garage_weighted_size"] = df["garage_cars"]*df["garage_area"]
    
    # create a total fireplace score
    df["fireplace_score"] = df["fireplaces"]*df["fireplace_qu"]
    
    # create a total kitchen score
    df["kitchen_score"] = df["kitchen_qual"]*df["kitchen_abvgr"]
    
    # create a combined basement score
    df["bsmt_score"] = df["bsmt_qual"]+ df["bsmt_cond"] + df["bsmt_exposure"] + \
                        df["bsmtfin_type_1"]+df["bsmtfin_type_2"]
    
    # create a combined exterior score
    df["exter_score"] = df["exter_qual"] + df["exter_cond"]
    
    # create a total_bathrooms count
    df["total_bathrooms"] = df["bsmt_half_bath"]*0.5 + df["half_bath"]*0.5 + \
                            df["full_bath"] + df["bsmt_full_bath"]
    
    # create total porch size score
    df["total_porch_size"] = df["screen_porch"] + df["3ssn_porch"] + df["enclosed_porch"] + \
                            df["open_porch_sf"] + df["wood_deck_sf"]
    
    # columns to drop
    removal_list = ["garage_area","garage_cars","pool_qc","pool_area","wood_deck_sf",\
                    "open_porch_sf","enclosed_porch","3ssn_porch","screen_porch",\
                    "garage_finish","garage_qual","garage_cond","paved_drive","fireplace_qu",\
                    "fireplaces","kitchen_abvgr","kitchen_qual","bsmt_full_bath","full_bath",\
                    "half_bath","bsmt_half_bath","exter_cond","exter_qual",'bsmtfin_sf_2',\
                    'bsmtfin_sf_1','bsmt_unf_sf',"bsmt_qual","bsmt_cond","bsmt_exposure",\
                    "bsmtfin_type_1","bsmtfin_type_2","garage_yr_blt"]
    
    df = df.drop(removal_list, axis=1, inplace=True)
```

```python
# ------------------------------ CREATE FUNCTION FOR ONE HOT ENCODING ------------------------------

def adding_dummies(df):
    df = pd.get_dummies(data=df, drop_first=True) 
    return df
```

```python
# apply functions to training set

convert_values(train)
combine_and_drop(train)
train = adding_dummies(train)
```

```python
# moving saleprice column to end of df
train["z_sale_price"] = train["saleprice"]
train.drop(["saleprice"], axis=1, inplace=True)
train.head()
```

```python
# rearrange columns according to names

train.sort_index(axis=1, inplace=True)
train.head()
```

```python
test.shape
```

```python
train.shape
```

### Train/Test Split

```python
X = train.iloc[:,:-1] # df excludes saleprice
y = train["z_sale_price"]
print(X.shape)
print(y.shape)

# 25-75 train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
```

```python
ss = StandardScaler()
ss.fit(X_train) 

X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

## Modeling
<a id='model'></a>

### 1. Linear Regression

```python
# establishing baseline scores

base_df= pd.DataFrame(y)
base_df['naivepred']= np.mean(y)
base_df.head()
```

```python
n_RMSE = np.sqrt(mean_squared_error(base_df['z_sale_price'], base_df['naivepred']))
print("NAIVE ROOT MEAN SQUARE ERROR: {}".format(n_RMSE))
```

```python
# linear regression model

lr = LinearRegression()
kf = KFold(n_splits = 10, shuffle = True, random_state = 42)
lr_score = cross_val_score(lr, X_train, y_train, cv=kf)
print("LINEAR REGRESSION SCORES FOR TRAINING SET\n-----------------------------------------\n")
print("LR CROSS-VALIDATED R-SQUARE SCORES: {}\n\nLR CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(lr_score,lr_score.mean()),"\n\n")


lr_score = cross_val_score(lr, X_test, y_test, cv=kf)
print("LINEAR REGRESSION SCORES FOR TEST SET\n-----------------------------------------\n")
print("LR CROSS-VALIDATED R-SQUARE SCORES: {}\n\nLR CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(lr_score,lr_score.mean()))
```

The mean R^2 is extremely negative. The linear regression model performs worse on the test set i.e. there is probably overfitting in the training set and the redundant variables are affecting the coefficients.

### 2. Ridge Regression

```python
# ridge

ridge_alphas = np.logspace(0, 5, 200) 
kf = KFold(n_splits = 10, shuffle = True, random_state = 42)
optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=kf)
optimal_ridge.fit(X_train, y_train)

print(optimal_ridge.alpha_)
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, X_train, y_train, cv=10)
print("RIDGE REGRESSION SCORES FOR TRAINING SET\n-----------------------------------------\n")
print("RIDGE CROSS-VALIDATED R-SQUARE SCORES: {}\n\nRIDGE CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(ridge_scores,np.mean(ridge_scores)),"\n\n")

ridge_scores = cross_val_score(ridge, X_test, y_test, cv=10)
print("RIDGE REGRESSION SCORES FOR TEST SET\n-----------------------------------------\n")
print("RIDGE CROSS-VALIDATED R-SQUARE SCORES: {}\n\nRIDGE CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(ridge_scores,np.mean(ridge_scores)))
```

The results with ridge regression is much better than linear regression. The score for the training set is slightly better than the test set i.e. it's showing potential to generalise well.

### 3. Laso Regression

```python
optimal_lasso = LassoCV(n_alphas=500, cv=10)
optimal_lasso.fit(X_train, y_train)

print (optimal_lasso.alpha_)
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)
kf = KFold(n_splits = 10, shuffle=True, random_state = 42)
lasso_scores = cross_val_score(lasso, X_train, y_train, cv=kf)
print("LASSO REGRESSION SCORES FOR TRAINING SET\n-----------------------------------------\n")
print("LASSO CROSS-VALIDATED R-SQUARE SCORES: {}\n\nLASSO CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(lasso_scores,np.mean(lasso_scores)),"\n\n")


lasso_scores = cross_val_score(lasso, X_test, y_test, cv=kf)
print("LASSO REGRESSION SCORES FOR TEST SET\n-----------------------------------------\n")
print("LASSO CROSS-VALIDATED R-SQUARE SCORES: {}\n\nLASSO CROSS-VALIDATED MEAN R-SQUARE SCORE: {}".
      format(lasso_scores,np.mean(lasso_scores)))
```

```python
# exploring lasso coefficients

lasso.fit(X_train, y_train)
lasso_coefs = pd.DataFrame({'variable':X.columns,
                            'coef':lasso.coef_,
                            'abs_coef':np.abs(lasso.coef_)})

lasso_coefs.sort_values('abs_coef', inplace=True, ascending=False)

lasso_coefs.head(30) #limit to 30
```

```python
# exploring negative lasso coefficients

lasso_coefs.loc[lasso_coefs['coef'] < 0,:].head(20) # limit to 20
```

```python
print ('Percent variables with zero coef:', np.sum((lasso.coef_ == 0))/float(X.shape[0]))
```

The lasso model performed similar to the ridge model. The lasso coefficients above also shows us features which would add value to a house and also featuers that would devalue a house e.g. too many bedrooms, mansard roofstyle and townhouses.

### 4. Elastic Net

```python
l1_ratios = np.linspace(0.01, 1.0, 25)
kf = KFold(n_splits = 10, shuffle=True, random_state = 42)
optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=kf)
optimal_enet.fit(X_train, y_train)

print (optimal_enet.alpha_)
print (optimal_enet.l1_ratio_)
```

```python
enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)
kf = KFold(n_splits = 10, shuffle=True, random_state = 42)
enet_scores = cross_val_score(enet, X_train, y_train, cv=kf)

print("E-NET REGRESSION SCORES FOR TRAINING SET\n-----------------------------------------\n")
print("E-NET CROSS-VALIDATED R-SQUARE SCORES: {}\n\nE-NET CROSS-VALIDATED MEAN R-SQUARE SCORE: {}"
      .format(enet_scores,np.mean(enet_scores)),"\n\n")

enet_scores = cross_val_score(enet, X_test, y_test, cv=kf)

print("E-NET REGRESSION SCORES FOR TEST SET\n-----------------------------------------\n")
print("E-NET CROSS-VALIDATED R-SQUARE SCORES: {}\n\nE-NET CROSS-VALIDATED MEAN R-SQUARE SCORE: {}"
      .format(enet_scores,np.mean(enet_scores)))
```

As we used a L1 Ratio of 1, this elastic net model is essentially a lasso model. Hence the similar results.

## Summary & Recommendations
<a id='sum'></a>

#### Location
The phrase ["location, location, location"](https://www.thebalance.com/what-location-means-in-real-estate-1798766) is still valid when it comes to property. Neighborhoods do play a part in the price of a house. In Ames, the most desirable neighbourhoods appear to be: 
- Northridge Heights: located in the Gilbert School District and close to Ada Hayden Lake and recreation area, Somerset Village restaurants and shopping, and North Grand Mall
- Stone Brook: 5 minute drive to Iowa State University, preferred neighborhood for students and faculty staff
- Northridge: situated close to three public schools, Somerset Village restaurants and shopping and North Grand Mall 

However, location is not all of it and there are other features that add value to a home.

#### House Type


- Top features: Above ground living area, newer houses, overall quality and condition. These are expected but is there more? 


- Basement score: This was an engineered feature where we combined all the different features of the basement e.g. quality, condition, exposure, rating of the basement. To raise home sale prices, homeowners can consider upgrading their basements.


- Surprising features: Garage and fireplaces. Upon doing some research, the reason for why fireplace(s) is an important feature in a house could be because [Iowa has one of the worst winters in the US](https://khak.com/iowa-ranked-among-states-with-most-miserable-winters/).


- Features to avoid: Townhouses and too many bed rooms. Although price genearally increases with the size of the house, it is negatively correlated with the number of bedrooms.

#### Can this model be generalized with other cities?

- As Ames is a college town where [half of its population are studens from Iowa State University](https://en.wikipedia.org/wiki/Ames,_Iowa), our model might only generalise well to similar [college towns in the US](https://en.wikipedia.org/wiki/List_of_college_towns#College_towns_in_the_United_States).
- The dataset we've used for this model is collected between 2006-2010 which coincides with the [subprime mortgage crisis](https://en.wikipedia.org/wiki/Subprime_mortgage_crisis). There was a large decline in housing prices which means that the data may not generalise well to other periods of time.
